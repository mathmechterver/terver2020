## Линейность математического ожидания

1. $E(a\xi) = \sum(a\xi_i)P(\xi = \xi_i) = a \sum(\xi_i)P(\xi = \xi_i) = aE\xi$

2. $E(\xi + \eta) 
=\\ \sum\limits_{x \in X_{\xi + \eta}}xP(\xi + \eta = x) 
=\\ \sum\limits_{s \in X_{\xi}, t \in X_{\eta}} (s+t) P(\xi =s, \eta = t) 
=\\ \sum\limits_{s \in X_{\xi}, t \in X_{\eta}} s P(\xi =s, \eta = t) + \sum\limits_{s \in X_{\xi}, t \in X_{\eta}} t P(\xi =s, \eta = t) 
=\\ \sum\limits_{s \in X_{\xi}, t \in X_{\eta}} s P(\eta = t |\xi =s)P(\xi =s) + \sum\limits_{s \in X_{\xi}, t \in X_{\eta}} t P(\xi =s| \eta = t)P(\eta = t) 
=\\ {\sum\limits_{s \in X_{\xi}} \sum\limits_{t \in X_{\eta}}}  s P(\eta = t |\xi =s)P(\xi =s) + {\sum\limits_{s \in X_{\xi}} \sum\limits_{t \in X_{\eta}}} t P(\xi =s| \eta = t)P(\eta = t) 
=\\{\sum\limits_{s \in X_{\xi}}}  sP(\xi =s)(\sum\limits_{t \in X_{\eta}}P(\eta = t |\xi =s)) + {\sum\limits_{t \in X_{\eta}}} t P(\eta = t)(\sum\limits_{s \in X_{\xi}} P(\xi =s| \eta = t)) 
=\\{\sum\limits_{s \in X_{\xi}}}  sP(\xi =s) + {\sum\limits_{t \in X_{\eta}}} t P(\eta = t) =\\ E \xi + E \eta$

## Задачи на пару

**1.** Пусть $\xi$ и $\eta$ - независимые случайные величины. Доказать, что $E\xi\eta = E \xi E \eta$.

**2.**
В каком случае $D(\xi + \mu) = D\xi + D\mu$? А в каком не равно? На какую величину они будут отличаться в общем случае? 

**3.**
Докажите свойства ковариаций
- $cov(\xi, \mu) = cov(\mu, \xi)$
- $cov(\alpha\xi, \mu) = \alpha cov(\xi, \mu)$
- $cov(\xi_1 + \xi_2, \mu) = cov(\xi_1, \mu) + cov(\xi_2, \mu)$

**4.**
Двумерное распределение пары целочисленных случайных величин $\xi$ и $\eta$ задаётся с помощью таблицы, где в пересечении столбца $\xi = i$ и строки $\eta = j$ находится вероятность $P\lbrace{\xi = i, \eta = j\rbrace}$. 

|$\xi$ & $\eta$     |$\xi$ $=$ $-1$    |$\xi$ $=$ $0$|$\xi$ $=$ $1$|
|-------------------|------------------|---------|---------|
|  $\eta$ $=$ $-1$  |$\frac{1}{8}$    |$\frac{1}{12}$ |$\frac{7}{24}$|
|  $\eta$ $=$ $1$  |$\frac{5}{24}$    |$\frac{1}{6}$ |$\frac{1}{8}$|

Найти:

- Дисперсию: $D(\xi), D(\eta)$,
- Ковариацию:  $Cov(\xi, \eta) $,
- Корреляцию: $\rho$ $(\xi, \eta)$

**5.**
Пусть в урне содержится $N$ шаров, из них $a$ белых, $b$ черных и $c$ красных, $a + b + c = N$. Вынимается $n$ шаров, и пусть среди них $\xi$ белых и $\eta$ черных. 

- Показать, что если совершаемый выбор –– это «выбор
с возвращением», то $Cov(\xi, \eta) = -npq$, где $p = \frac{a}{N}$ и $q = \frac{b}{N}$. Если же производится «выбор без возвращения», то $Cov(\xi, \eta) = -npq\frac{N - n}{N - 1}$.

- Показать также, что в обоих случаях корреляция $\rho(\xi, \eta) = -\sqrt{\frac{pq}{(1 - p)(1 - q)}}$.
