## Основные формулы

Геометрическое распределение
$$P(k)= p(1-p)^k$$

Биноминальное распределение
$$P_n(k) = C_n^kp^k(1-p)^{n-k}$$

Распределение Пуассона
$$P(k) = \frac{\lambda^{k}e^{-\lambda}}{k!}$$
