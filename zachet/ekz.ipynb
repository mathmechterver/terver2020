### Рогачёв (0)

1. Пусть в углах кубика сидят жуки. Жуки могут пойти по любому из 3-х ребер кубика, которые им доступны, с равной вероятностью. Пусть они все вместе начали движение. Какова вероятность, что никто не встретится?

2. На день рождения Дианы мы принесли торт с N свечами. Когда мы пытаемся задуть свечи, мы точно задуваем одну свечу, а количество оставшихся гореть свечей распределено равновероятно. (от 0 до N-1). При следующем задуве ситуация аналогичная. Вопрос: найти матожидание числа задуваний, которое должна сделать Диана.

С вероятностью p на середине тропинки может проснутся злобный маг, который построит барьер перпендикулярный тропе и по длине равный тропе. В этом случае нужно будет обойти данный барьер. Какой минимальный путь суммарно пройдет за N дней наш рыцарь?
